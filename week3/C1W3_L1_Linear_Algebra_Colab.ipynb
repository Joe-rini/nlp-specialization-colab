# üßÆ Part 2: Vector Addition and Scalar Multiplication

```python
a = np.array([1, 2])
b = np.array([3, 4])

print("a + b:", a + b)
print("3 * a:", 3 * a)
```

```python
origin = [0], [0]  # origin point
plt.quiver(*origin, a[0], a[1], color='blue', scale=1, scale_units='xy', angles='xy', label='a')
plt.quiver(*origin, b[0], b[1], color='green', scale=1, scale_units='xy', angles='xy', label='b')
plt.quiver(*origin, (a+b)[0], (a+b)[1], color='red', scale=1, scale_units='xy', angles='xy', label='a + b')
plt.quiver(*origin, (3*a)[0], (a+b)[1], color='orange', scale=1, scale_units='xy', angles='xy', label='3a')
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.grid(True)
plt.legend()
plt.title("Vector Addition: a + b, 3a")
plt.show()
```

# üîç Part 3: Dot Product and Magnitude
The dot product measures how similar two vectors are. It's foundational for cosine similarity.

```python
u = np.array([1, 2])
v = np.array([3, 4])

print("Dot product:", np.dot(u, v))
print("Magnitude of u:", np.linalg.norm(u))
print("Magnitude of v:", np.linalg.norm(v))
```

‚úÖ Dot Product
Formula: u ¬∑ v = u‚ÇÅ¬∑v‚ÇÅ + u‚ÇÇ¬∑v‚ÇÇ + ...

Meaning: Measures how much two vectors point in the same direction.

- If large & positive: Vectors point in similar directions.
- If 0: Vectors are orthogonal (90¬∞ apart).
- If negative: Vectors point in opposite directions.

‚úÖ Magnitude (‚Äñu‚Äñ or ‚Äñv‚Äñ)
- Formula: ‚Äñu‚Äñ = sqrt(u‚ÇÅ¬≤ + u‚ÇÇ¬≤ + ‚Ä¶)
- Meaning: The "length" of the vector in space.
- Used to normalize vectors or compute angles.

‚úÖ Cosine Similarity
Formula: cos(Œ∏) = (u ¬∑ v) / (‚Äñu‚Äñ ¬∑ ‚Äñv‚Äñ)

Meaning: A number from -1 to 1 that tells how similar two directions are.

- 1 = same direction
- 0 = 90¬∞ apart
- -1 = opposite directions

Let's plot it

```python
# Origin point for both vectors
origin = np.array([[0, 0], [0, 0]])

# Calculate dot product and angle
dot = np.dot(u, v)
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)
cos_sim = dot / (norm_u * norm_v)
angle_deg = np.degrees(np.arccos(cos_sim))

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(*origin, [u[0], v[0]], [u[1], v[1]], color=["blue", "green"], angles='xy', scale_units='xy', scale=1)

# Vector labels
plt.text(u[0] * 1.05, u[1] * 1.05, "u", fontsize=12, color="blue")
plt.text(v[0] * 1.05, v[1] * 1.05, "v", fontsize=12, color="green")

# Dot product annotation
plt.text(0.5, 4.5, f"Dot Product = {dot}", fontsize=12, color="black")
plt.text(0.5, 4.2, f"Angle ‚âà {angle_deg:.2f}¬∞", fontsize=12, color="gray")

# Plot styling
plt.title("Vectors u and v with Dot Product")
plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.gca().set_aspect('equal')
plt.show()
```

üí° **Cosine Similarity** = `dot(u, v) / (||u|| * ||v||)`

# üåç Part 4: Real Example - Embedding Similarity

```python
king = np.array([0.8, 0.1, 0.3])
queen = np.array([0.7, 0.1, 0.4])
apple = np.array([0.2, 0.9, 0.3])
```

```python
def cosine_similarity(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

print("king vs queen:", cosine_similarity(king, queen))
print("king vs apple:", cosine_similarity(king, apple))
```

üëÜ The model thinks **king** and **queen** are more alike than **king** and **apple**, just from numbers.
